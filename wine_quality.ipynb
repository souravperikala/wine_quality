```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import preprocessing
```

```python
mydata=pd.read_csv('winequality_red.csv')
```

```python
data=mydata[:]
```

```python
data['quality'].replace([1,2,3,4,5,6,7,8,9],[0,0,0,0,0,0,1,1,1],inplace=True)
```

# normalising the data variables

```python
X=data.loc[:,['alcohol','sulphates','citric acid','fixed acidity']]
y=labels
```

```python
from sklearn import preprocessing
minmax=preprocessing.MinMaxScaler(feature_range=(0,1))
minmax.fit(X).transform(X)
```

# KNN model

```python
from sklearn import model_selection,neighbors
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
clf=neighbors.KNeighborsClassifier()
clf.fit(X_train,y_train)
```

```python
#prediction on test data
y_pred=clf.predict(X_test)
print("predicted quality")
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,recall_score,roc_auc_score,confusion_matrix
print("\n Accuracy score:%f"%(accuracy_score(y_test,y_pred)*100))
print("recall_score:%f"%(recall_score(y_test,y_pred)*100))
print("ROC score:%f\n"%(roc_auc_score(y_test,y_pred)*100))
print(confusion_matrix(y_test,y_pred))
```

# visualising the probability

```python
probas=clf.predict_proba(X_test)
import matplotlib.pyplot as plt
plt.figure(dpi=150)
plt.hist(probas,bins=20)
plt.title('classification probabilities')
plt.xlabel('probability')
plt.ylabel('# of instances')
plt.xlim([0.5,1.0])
plt.legend(y_test)
plt.show()
```

```python
#evaluation of matrics
X_train_std=minmax.fit_transform(X_train)
X_test_std=minmax.fit_transform(X_test)
from sklearn.model_selection import cross_val_score,cross_val_predict
#print cross validation for knn
clf_acc=cross_val_score(clf,X_train_std,y_train,cv=3,scoring="accuracy",n_jobs=-1)
```

```python
clf_proba=cross_val_predict(clf,X_train_std,y_train,cv=3,method='predict_proba')
clf_scores=clf_proba[:,1] #get score of positive class
```
